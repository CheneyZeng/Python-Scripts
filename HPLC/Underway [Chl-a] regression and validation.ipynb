```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
import sklearn.model_selection as sms
import pandas as pd

%matplotlib inline
```

```python
xl = pd.ExcelFile("C:/computer_harddrive/matching.xlsx")
df = xl.parse("python")
df.head()
```

```python
sns.set_style("ticks")
fig = plt.figure(figsize=(6, 4))

x = df['alh']
y = df['chl']

# Extend x data to contain another row vector of 1s
X = np.vstack([x, np.ones(len(x))]).T

for i in range(0, 1000):
    sample_index = np.random.choice(range(0, len(y)), len(y))

    X_samples = X[sample_index]
    y_samples = y[sample_index]    

    lr = LinearRegression()
    lr.fit(X_samples, y_samples)
    plt.plot(x, lr.predict(X), color='grey', alpha=0.01, zorder=1)

lr = LinearRegression()
lr.fit(X, y)
plt.plot(x, lr.predict(X), color='black', linestyle ='--',zorder=3)
sr = "y=%.3fx+%.3f" % (lr.coef_[0], lr.intercept_)
plt.scatter(x,y, marker='o', color='black', zorder=2)
#plt.suptitle(sr,fontsize=20) 

plt.xlim([0, 0.2])
plt.ylim([0, 30])
plt.tight_layout()
plt.show()
```
