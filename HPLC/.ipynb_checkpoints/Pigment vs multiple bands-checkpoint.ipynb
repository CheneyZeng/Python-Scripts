```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
#read txt 
## here's a question left, what is the spec? aph or ap?
spec = np.genfromtxt('C:/computer_harddrive/oceanus/HPLC/pigment_hplc/spec.txt',
                    delimiter='\t')
hplc = np.genfromtxt('C:/computer_harddrive/oceanus/HPLC/pigment_hplc/pig.txt',
                    delimiter='\t')
pig_name = (['Tot_Chl_a', 'Fuco', 'Perid', 'Hex_fuco',
             'Allo', 'But_fuco', 'Tot_Chl_b', 'Zea', 'DP'])
pig_name
```

```python
hplc
```

```python
# transpose for cor matrix
x = np.transpose(spec)
y = np.transpose(hplc)
y.shape
```

```python
x_mean = x.mean(axis=1)
y_mean = y.mean(axis=1)
n = x.shape[1]

x_std = x.std(axis=1, ddof=n-1)
y_std = y.std(axis=1, ddof=n-1)

cov = np.dot(x, y.T) - n * np.dot(x_mean[:, np.newaxis], 
                                  y_mean[np.newaxis, :])

cor_coef = cov / np.dot(x_std[:, np.newaxis], y_std[np.newaxis, :])

cor_coef.shape
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3, sharex=True,figsize=(10,8))


for i in range(0, cor_coef.shape[1]):
    ax = axes[i//3, i%3]
    ax.plot(cor_coef[:, i])
    ax.set_title(pig_name[i] )

plt.tight_layout()
x = [0, 49, 99, 149]
sx = [400, 500, 600, 700]
plt.xticks(x, sx)
plt.show()
```

```python
# pick up maximum band for each pigments here and make a linear regression
cor_coef.max(axis=0)
index = np.argmax(cor_coef[20:,:], axis=0)
```

```python
# Linear regression for every feature pigments
pos = (spec[:, 1] > 0) & (spec[:, 1] < 0.4)
fig, axes = plt.subplots(nrows=3,ncols=3,figsize=(10,8))

for i in range(0, cor_coef.shape[1]):
    x = np.log10(hplc[pos,i])
    y = np.log10(spec[pos,index[i]+20])
    ax = axes[i//3, i%3]
    ax.scatter(x, y);
    ax.set_title(str(i))

plt.tight_layout()
plt.show()
```

### From the scatter plot, the problem is linear regression is unevenly distributed, does that mean it is not good to use linear regression?

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
fig, axes = plt.subplots(nrows=3,ncols=3, figsize=(10,8))

for i in range(0, cor_coef.shape[1]):
    pos = (spec[:, 1] > 0) & (spec[:, 1] < 0.4) & (spec[:, index[i]+20] > 0)
    x = np.log10(hplc[pos,i])
    y = np.log10(spec[pos,index[i]+20])
    model.fit(x[:, np.newaxis], y)
    xfit = np.linspace(-3, 0.1, 2)
    yfit = model.predict(xfit[:, np.newaxis])

    ax = axes[i//3, i%3]
    ax.scatter(x, y)
    ax.plot(xfit, yfit)
    sr = "y=%.3fx+%.3f" % (model.coef_[0], model.intercept_)
    ax.set_title(sr) 

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10,8))

for j in range(0, cor_coef.shape[1]):
    # Create sample data 
    pos = (spec[:, 1] > 0) & (spec[:, 1] < 0.4) & (spec[:, index[j]+20] > 0)
    x = np.log(hplc[pos,j])
    y = np.log(spec[pos,index[j]+20])

    ax = axes[j//3, j%3]
    # Extend x data to contain another row vector of 1s
    X = np.vstack([x, np.ones(len(x))]).T

    for i in range(0, 1000):
        sample_index = np.random.choice(range(0, len(y)), len(y))

        X_samples = X[sample_index]
        y_samples = y[sample_index]    

        lr = LinearRegression()
        lr.fit(X_samples, y_samples)
        ax.plot(x, lr.predict(X), color='grey', alpha=0.2, zorder=1)

    ax.scatter(x,y, marker='o', color='orange', zorder=2)
    lr = LinearRegression()
    lr.fit(X, y)
    ax.plot(x, lr.predict(X), color='red', zorder=3)
    sr = pig_name[j] + ":y=%.3fx+%.3f" % (lr.coef_[0], lr.intercept_)
    ax.set_title(sr) 

plt.tight_layout()
plt.show()
```
