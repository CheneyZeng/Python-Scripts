```python
#get micro% estimation
thplc['dp']=(1.41*thplc['fuco']+1.41*thplc['perid'] +1.27*thplc['hexx']+
             0.6*thplc['allo']+0.35*thplc['but']+1.01*thplc['chlb']+0.86*thplc['zea'])

thplc['micro']=(1.41*thplc['fuco']+1.41*thplc['perid'])/thplc['dp']
thplc['diatom']=(1.41*thplc['fuco'])/thplc['dp']
thplc['dino']=(1.41*thplc['perid'])/thplc['dp']

thplc['nano']=(0.5*1.27*thplc['hexx']+0.6*thplc['allo']+0.35*thplc['but']+1.01*thplc['chlb']
              )/thplc['dp']
thplc['green_algae']=(1.01*thplc['chlb'])/thplc['dp']
thplc['prymn']=(0.5*1.27*thplc['hexx']+0.35*thplc['but'])/thplc['dp']

thplc['pico']=(0.5*1.27*thplc['hexx']+0.86*thplc['zea'])/thplc['dp']
thplc['prokary']=(0.86*thplc['zea'])/thplc['dp']
thplc['pico_eu']=(0.5*1.27*thplc['hexx'])/thplc['dp']

#thplc = thplc.replace(0.000000, np.nan)
thplc.head()
```

```python
# Hirata's model
y = thplc['chla']
x = np.log10(y)
thplc['micro_chl'] = 1./ (0.9117 + np.exp (-2.7330* x + 0.4003)) #Micro
thplc['diatom_chl'] = 1./ (1.3272 + np.exp (-3.9828* x + 0.1953)) #Diatom
thplc['dino_chl'] = thplc['micro_chl'] - thplc['diatom_chl'] #Dino

thplc['pico_chl'] = -1./ (0.1529 + np.exp (1.0306* x - 1.5576)) - 1.8597*x +2.9954
thplc['prokary_chl'] = (0.0067/0.6154/y)*np.exp(-19.5190*((x + 0.9643)**2)/(0.6154**2)) + 0.1027*(x**2) \
    -0.1189*x +0.0626
thplc['pico_eu_chl'] = thplc['pico_chl'] - thplc['prokary_chl']

thplc['nano_chl'] = 1 - thplc['micro_chl'] - thplc['pico_chl']
thplc['green_algae_chl'] = (0.2490/y) * np.exp(-1.2621*(x-0.5523)**2)
thplc['prymn_chl'] = thplc['nano_chl'] - thplc['green_algae_chl']
```

```python
# curve fitting for MICRO
def func(x, a0, a1, a2):
  return 1./(a0+np.exp(a1*x+a2))

xdata = np.log10(ab['chla'].values)
ydata = ab['micro'].values

popt_micro, pcov = curve_fit(func, xdata, ydata,p0=(0.9117,-2.7330,0.4003))
sigma = np.sqrt([pcov[0,0], pcov[1,1], pcov[2,2]])
sigma[0]/popt_micro[0],sigma[1]/popt_micro[1],sigma[2]/popt_micro[2]
sigma[0],popt_micro[0],sigma[1],popt_micro[1],sigma[2],popt_micro[2]
```

```python
# curve fitting for DIATOM
def func(x, a0, a1, a2):
  return 1./ (a0 + np.exp (a1* x +a2 )) 

xdata = np.log10(ab['chla'].values)
ydata = ab['diatom'].values

popt_diatom, pcov = curve_fit(func, xdata, ydata,p0=(1.3272,-3.9828,0.1953))
sigma = np.sqrt([pcov[0,0], pcov[1,1], pcov[2,2]])
sigma[0]/popt_diatom[0],sigma[1]/popt_diatom[1],sigma[2]/popt_diatom[2]
sigma[0],popt_diatom[0],sigma[1],popt_diatom[1],sigma[2],popt_diatom[2]
```

```python
# curve fitting for PICO
def func(x, a0, a1, a2, a3, a4):
  return -1./ (a0 + np.exp (a1* x + a2)) + a3*x +a4

xdata = np.log10(ab['chla'].values)
ydata = ab['pico'].values

popt_pico, pcov = curve_fit(func, xdata, ydata,p0=(0.1529,1.0306,-1.5576,-1.8597,2.9954),maxfev=10000)
sigma = np.sqrt([pcov[0,0], pcov[1,1], pcov[2,2],pcov[3,3],pcov[4,4]])
(sigma[0]/popt_pico[0],sigma[1]/popt_pico[1],sigma[2]/popt_pico[2],
sigma[3]/popt_pico[3],sigma[4]/popt_pico[4])
(sigma[0],popt_pico[0],sigma[1],popt_pico[1],sigma[2],popt_pico[2],
sigma[3],popt_pico[3],sigma[4],popt_pico[4])
```

```python
# curve fitting for prokary
def func(x, a0, a1, a2, a3, a4, a5, a6):
  return ((a0/a1/(10**x))*np.exp(a2*((x + a3)**2)/(a1**2)) + a4*(x**2) +a5*x +a6)

xdata = np.log10(ab['chla'].values)
ydata = ab['prokary'].values

popt_prokary,pcov = curve_fit(func, xdata, ydata,p0=(0.0067,0.6154,-19.5190,0.9643,0.1027,-0.1189,0.0626),
                              maxfev=10000,method='trf')
sigma = np.sqrt([pcov[0,0], pcov[1,1], pcov[2,2],pcov[3,3],pcov[4,4],pcov[5,5],pcov[6,6]])
(sigma[0]/popt_prokary[0],sigma[1]/popt_prokary[1],sigma[2]/popt_prokary[2],
sigma[3]/popt_prokary[3],sigma[4]/popt_prokary[4],sigma[5]/popt_prokary[5],
sigma[6]/popt_prokary[6])
```

```python
# curve fitting for greenalgae
def func(x, a0, a1, a2):
  return (a0/(10**x)) * np.exp(a1*(x+a2)**2)

xdata = np.log10(ab['chla'].values)
ydata = ab['green_algae'].values

popt_ga, pcov = curve_fit(func, xdata, ydata,p0=(0.2490,-1.2621,-0.5523),maxfev=120000,method='trf')
sigma = np.sqrt([pcov[0,0], pcov[1,1], pcov[2,2]])
sigma[0]/popt_ga[0],sigma[1]/popt_ga[1],sigma[2]/popt_ga[2]
```

```python
# Hirata's model
y = np.arange(0.01, 100, 0.1)
x = np.log10(y)
micro_new = 1./ (popt_micro[0] + np.exp (popt_micro[1]* x + popt_micro[2])) #Micro
diatom_new = 1./ (popt_diatom[0] + np.exp (popt_diatom[1]* x + popt_diatom[2])) #Diatom
dino_new = micro_new - diatom_new #Dino

pico_new = -1./ (popt_pico[0] + np.exp (popt_pico[1]* x + popt_pico[2])) + popt_pico[3]*x +popt_pico[4]
prokaryotes_new = (popt_prokary[0]/popt_prokary[1]/y)*np.exp(popt_prokary[2]*((x + popt_prokary[3])**2)/
                                                             (popt_prokary[1]**2)) + popt_prokary[4]*(x**2) \
    +popt_prokary[5]*x +popt_prokary[6]
picoeukaryotes_new = pico_new - prokaryotes_new

nano_new = 1 - micro_new - pico_new
greenalgae_new = (popt_ga[0]/y) * np.exp(popt_ga[1]*(x+popt_ga[2])**2)
prymnesiophytes_new = nano_new - greenalgae_new
```
