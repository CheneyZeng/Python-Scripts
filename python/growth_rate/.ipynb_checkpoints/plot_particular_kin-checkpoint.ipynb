```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
xl = pd.ExcelFile("C:/Users/user/Documents/Python_Scripts/growth_rate/ALL_DATA_for_matlab.xlsx")
df = xl.parse("plot_data")
df = df.replace('Flag','Aflag')
df.head()
```

```python
proj = df['Project'].unique()
proj 
```

```python
sns.set_style("ticks")
# select each of the project and plot them in figure ---> check
# group 'Temp' by ranges

sub = df[(df['Project']==proj[0]) & (df['Cell type1']=='Diatom')]
c = "purple"
bp = sub['kin/SA model'].groupby(pd.cut(sub["Temp"], np.arange(-2, 30, 2))
                            ).agg([np.nanmean,np.nanstd]).transpose().boxplot(vert=False, patch_artist=True,
                                                                             )
plt.setp(bp["boxes"], facecolor=c, color=c)
plt.setp(bp["whiskers"], color=c)

plt.xscale('log')
plt.xlim([10e-9,10e-13])
plt.title('Diatom')
plt.show()
```

```python
sns.set_style("ticks")
# select each of the project and plot them in figure ---> check
# group 'Temp' by ranges
plist = [0,2,3,4,5,6,7,8]
typelist = ['Diatom', 'Aflag', 'Apico']
colorlist = ["purple","red","orange","plum","navy","tomato","skyblue","yellowgreen"]

for color, p in zip(colorlist, plist):
    sub = df[(df['Project']==proj[p]) & (df['Cell type1']==typelist[0])]
    c = color
    bp = sub['kin/SA model'].groupby(pd.cut(sub["Temp"], np.arange(-2, 0.1, 30))
                                    ).agg([np.nanmean,np.nanstd]).transpose().boxplot(vert=False, patch_artist=True, )
    plt.setp(bp["boxes"], facecolor=c, color=c)
    plt.setp(bp["whiskers"], color=c)
    plt.hold(True)

plt.xscale('log')
plt.xlim([10e-9,10e-13])
plt.title(typelist[0])
plt.legend((proj[plist]))
plt.ylabel("Temperature")
plt.xlabel("kin Practical / SA")
plt.show()
```
